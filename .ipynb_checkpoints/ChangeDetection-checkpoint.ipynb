```python
import pandas as pd
import seaborn as sns ; sns.set()
import matplotlib.pyplot as plt
import csv
%matplotlib inline
```

```python
df_origCleaned = pd.read_csv('Cleaned.csv')
df_orig200 = df_origCleaned.head(10)
LinePlot = sns.lineplot(data=df_orig200, x='TimeUnix', y='x')
```

```python
# plotting all x, y, z dimensions togather.
from matplotlib.pyplot import figure
fig = plt.figure()
fig.set_size_inches(18, 5, forward=True)
ax = plt.axes()
ax.plot(df_orig200["TimeUnix"], df_orig200["x"])
ax.plot(df_orig200["TimeUnix"], df_orig200["y"])
ax.plot(df_orig200["TimeUnix"], df_orig200["z"])
ax.legend()
ax.can_zoom()
ax.get_data_ratio();
```
